```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

# import data visualization packages
```

```python
wine = pd.read_excel('wine_datason.xlsx') # read the data 
```

```python
wine.info() # understand the data  
```

```python
wine.isnull().sum().sum() # no missing data
```

```python
sns.set() # set the style
```

```python
sns.set_style("dark") # set the background color
sns.displot(wine, x= "Sugar", height=7) # create histogram graph of sugar and its height
plt.title("The Distribution of Sugar in Wines", fontweight="bold", fontsize=15, loc="center") # set the title, its font weight, font size and location
plt.xlabel("Sugar") # set the x-axis name
plt.ylabel("Count") # set the y-axis name
plt.show() # show the final version of the histogram graph
```

```python
plt.figure(figsize=(50,50)) # arrange the size of the clustered bar graph
wine.groupby("Wine_Type").aggregate({"Quality_Score":"mean"}).plot(kind = "barh", color = "steelblue") # categorize wine types by using groupedby and compare their quality scores
plt.title("Wine Types and Their Average Quality Scores", fontweight="bold", fontsize=15, loc="center") # set the title, its font weight, font size and location
plt.xlabel("Quality Score") # set the x-axis name
plt.ylabel("Wine Type") # set the y-axis name
plt.legend("", frameon=False)
plt.show() # show the final version of the bar graph
```

```python
plt.figure(figsize=(10,10)) #arrange the size of the clustered bar graph
sns.barplot(data = wine, x = "Wine_Type", 
            y = "Sugar",
            hue = "Color",
            palette = ['red','lightyellow','pink'])# set the axis and hue
plt.xlabel("Wine Type") # set the x-axis name
plt.ylabel("Sugar") # set the y-axis name
plt.title("The Average Sugar Amounts by Wine Colors for Wine Types", fontweight="bold", fontsize=15, loc="center") # set the title, its font weight, font size and location
plt.show() # show the final version of the clustered bar graph
```

```python
sns.catplot(data=wine, x="Color", 
            kind="count",
            color='steelblue',
            height=7, 
            palette = ['red','lightyellow','pink']) # set the axis and its size
plt.xlabel("Color") # set the x-axis name
plt.ylabel("Count") # set the y-axis name
plt.title('The Frequency Distribution of Color Types of Wines', fontweight="bold",fontsize=15, loc="center") # set the title, its font weight, font size and location
plt.show() # show the final version of the bar graph
```

```python
plt.figure(figsize = (10,10)) # arrange the size
sns.boxplot(data=wine, 
            x="Cholorides", 
            y="Sugar",
            palette = 'Paired') # set the axis and palette
plt.title('Sugar Amounts for Cholorides',fontsize = 15, fontweight = "bold", loc="center") # set the title, its font weight, font size and location
plt.xlabel('Cholorides') # set the x-axis name
plt.ylabel('Sugar') # set the y-axis name
plt.show() # show the final version of the box-plot
```

```python
plt.figure(figsize = (10,10)) # arrange the size
sns.scatterplot(data=wine, x="Alcohol", 
                y="Sugar",
                hue = 'Color', 
                palette = ['Red','Lightyellow','Pink'],
                s=100) # set the axis and palette
plt.title('The Relationship Between Alcohol and Sugar',fontsize = 15, fontweight = "bold", loc="center") # set the title, its font weight, font size and location
plt.xlabel('Alcohol') # set the x-axis name
plt.ylabel('Sugar') # set the y-axis name
plt.show() # show the final version of the scatter graph
```

```python
plt.figure(figsize = (13,13)) # arrange the size
sns.violinplot(data=wine, 
               x="Grape", 
               y="Total_Sulfur_Dioxide",
               palette = 'Paired') # set the axis and palette
plt.title('Total Sulfur Dioxide by Grape Types',fontsize = 15, fontweight = "bold", loc="center")  # set the title, its font weight, font size and location
plt.xlabel('Grape') # set the x-axis name
plt.ylabel('Total Sulfur Dioxide') # set the y-axis name
plt.show() # show the final version of the violin plot
```
